```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random numbers for the prior and posterior distributions
np.random.seed(42)  # For reproducibility

# Simulate prior distribution (probability of rain)
prior = np.random.beta(a=3, b=5, size=100)

# Simulate likelihood (carrying umbrella given rain and not rain)
likelihood_rain = np.random.beta(a=9, b=5, size=100)  # Adjusted beta distribution parameters for likelihood
likelihood_no_rain = np.random.beta(a=5, b=10, size=100)  # Adjusted beta distribution parameters for likelihood

# Simulate posterior distribution (updated probability based on observed data)
posterior = np.random.beta(a=5, b=3, size=100)

# Plot the distributions using seaborn
plt.figure(figsize=(12, 6))

# Plot prior distribution
#plot kernel density estimates, alpha is to make the colors 50% transparent for a better look
sns.kdeplot(prior, label='Prior Probability Density Function (PDF)', color='blue', fill=True, alpha=0.5)

# Plot likelihood distributions
sns.kdeplot(likelihood_rain, label='Likelihood (Rain)', color='green', fill=True, alpha=0.5)
sns.kdeplot(likelihood_no_rain, label='Likelihood (No Rain)', color='orange', fill=True, alpha=0.5)

# Plot posterior distribution
sns.kdeplot(posterior, label='Posterior PDF', color='red', fill=True, alpha=0.5)

# Add titles and labels
plt.title('Prior, Likelihood, and Posterior PDF')
plt.xlabel('Probability')
plt.ylabel('Density')
plt.legend()

# Set x-axis limits to ensure it's within [0, 1]
plt.xlim(0, 1)

# Show the plot
plt.show()
```
