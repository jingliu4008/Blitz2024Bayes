```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Hypothetical probabilities
prior_rain = 0.15
likelihood_umbrella_given_rain = 0.3
likelihood_umbrella_given_no_rain = 0.1
prior_no_rain = 1 - prior_rain

# Total probability of carrying an umbrella
p_umbrella = (likelihood_umbrella_given_rain * prior_rain) + (likelihood_umbrella_given_no_rain * prior_no_rain)

# Posterior probability of rain given umbrella
posterior_rain_given_umbrella = (likelihood_umbrella_given_rain * prior_rain) / p_umbrella
posterior_no_rain_given_umbrella = (likelihood_umbrella_given_no_rain * prior_no_rain) / p_umbrella

# Generate random samples for prior and posterior distributions
np.random.seed(42)
prior_samples = np.random.choice([1, 0], size=1000, p=[prior_rain, prior_no_rain])
posterior_samples = np.random.choice([1, 0], size=1000, p=[posterior_rain_given_umbrella, posterior_no_rain_given_umbrella])

# Plotting the prior and posterior distributions
plt.figure(figsize=(12, 6))

# Prior distribution
sns.kdeplot(prior_samples, label='Prior (Probability of Rain)', color='blue', fill=True, alpha=0.5)

# Posterior distribution
sns.kdeplot(posterior_samples, label='Posterior (Probability of Rain given Umbrella)', color='red', fill=True, alpha=0.5)

plt.title('Prior and Posterior Distributions')
plt.xlabel('Probability')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Hypothetical probabilities
prior_rain = 0.5
likelihood_umbrella_given_rain = 0.8
likelihood_umbrella_given_no_rain = 0.05
prior_no_rain = 1 - prior_rain

# Total probability of carrying an umbrella
p_umbrella = (likelihood_umbrella_given_rain * prior_rain) + (likelihood_umbrella_given_no_rain * prior_no_rain)

# Posterior probability of rain given umbrella
posterior_rain_given_umbrella = (likelihood_umbrella_given_rain * prior_rain) / p_umbrella
posterior_no_rain_given_umbrella = (likelihood_umbrella_given_no_rain * prior_no_rain) / p_umbrella

# Generate random samples for prior, likelihood, and posterior distributions
# for binary varibales like rain/no rain, umbrella/no umbrella:
np.random.seed(42)
prior_samples = np.random.choice([1, 0], size=1000, p=[prior_rain, prior_no_rain])
likelihood_samples_rain = np.random.choice([1, 0], size=1000, p=[likelihood_umbrella_given_rain, 1-likelihood_umbrella_given_rain])
likelihood_samples_no_rain = np.random.choice([1, 0], size=1000, p=[likelihood_umbrella_given_no_rain, 1-likelihood_umbrella_given_no_rain])
posterior_samples = np.random.choice([1, 0], size=1000, p=[posterior_rain_given_umbrella, posterior_no_rain_given_umbrella])

# Plotting the prior, likelihood, and posterior distributions
plt.figure(figsize=(12, 6))

# Prior distribution
sns.kdeplot(prior_samples, label='Prior (Probability of Rain)', color='blue', fill=True, alpha=0.5)

# Likelihood distribution
sns.kdeplot(likelihood_samples_rain, label='Likelihood (Umbrella given Rain)', color='green', fill=True, alpha=0.5)
sns.kdeplot(likelihood_samples_no_rain, label='Likelihood (Umbrella given No Rain)', color='orange', fill=True, alpha=0.5)

# Posterior distribution
sns.kdeplot(posterior_samples, label='Posterior (Probability of Rain given Umbrella)', color='red', fill=True, alpha=0.5)

plt.title('Prior, Likelihood, and Posterior Distributions')
plt.xlabel('Probability')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# can use other distributions other than normal distribution

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random numbers for the prior and posterior distributions
np.random.seed(42)  # For reproducibility

# Simulate prior distribution (probability of rain)
prior = np.random.beta(a=3, b=5, size=100)

# Simulate likelihood (carrying umbrella given rain and not rain)
likelihood_rain = np.random.beta(a=9, b=5, size=100)  # Adjusted beta distribution parameters for likelihood
likelihood_no_rain = np.random.beta(a=5, b=10, size=100)  # Adjusted beta distribution parameters for likelihood

# Simulate posterior distribution (updated probability based on observed data)
posterior = np.random.beta(a=5, b=3, size=100)

# Plot the distributions using seaborn
plt.figure(figsize=(12, 6))

# Plot prior distribution
#plot kernel density estimates, alpha is to make the colors 50% transparent for a better look
sns.kdeplot(prior, label='Prior Probability Density Function (PDF)', color='blue', fill=True, alpha=0.5)

# Plot likelihood distributions
sns.kdeplot(likelihood_rain, label='Likelihood (Rain)', color='green', fill=True, alpha=0.5)
sns.kdeplot(likelihood_no_rain, label='Likelihood (No Rain)', color='orange', fill=True, alpha=0.5)

# Plot posterior distribution
sns.kdeplot(posterior, label='Posterior PDF', color='red', fill=True, alpha=0.5)

# Add titles and labels
plt.title('Prior, Likelihood, and Posterior PDF')
plt.xlabel('Probability')
plt.ylabel('Density')
plt.legend()

# Set x-axis limits to ensure it's within [0, 1]
plt.xlim(0, 1)

# Show the plot
plt.show()
```
